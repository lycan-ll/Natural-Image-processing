```python

def IDF(corpus, term):
    document_count = sum(1 for doc in corpus if term in doc)
    if document_count == 0:
        return 0
    total_documents = len(corpus)
    return math.log(total_documents / document_count)
```

```python
high_freq_idf = [w for w in words if IDF([text1], w) > 1]
```

```python
words_text1 = [word.lower() for word in text1 if word.isalpha()]
words_text1 = sorted(set(words_text1))

# Calculate bigrams and trigrams for Text1
bigrams_text1 = list(nltk.bigrams(text1))
trigrams_text1 = list(nltk.trigrams(text1))

# Get the most common bigrams and trigrams for Text1
fdist_bigrams_text1 = nltk.FreqDist(bigrams_text1)
most_common_bigrams_text1 = fdist_bigrams_text1.most_common(10)

fdist_trigrams_text1 = nltk.FreqDist(trigrams_text1)
most_common_trigrams_text1 = fdist_trigrams_text1.most_common(5)

# Other calculations for Text1
num_words_length_gt_16_text1 = len([w for w in words_text1 if len(w) > 16])

fdist_words_text1 = nltk.FreqDist(words_text1)
num_words_freq_gt_500_text1 = len([w for w in words_text1 if fdist_words_text1[w] > 500])
num_words_ending_ed_text1 = len([w for w in words_text1 if w.endswith('ed')])

# Repeat the above process for Text2 and Text3

# Display the table
print("Text1")
print("10 frequently occurring Bigrams:")
for bigram, freq in most_common_bigrams_text1:
    print(bigram, "-", freq)

print("\n5 frequently occurring Trigrams:")
for trigram, freq in most_common_trigrams_text1:
    print(trigram, "-", freq)

print("\nNumber of words with length > 16:", num_words_length_gt_16_text1)
print("Number of words with frequency > 500:", num_words_freq_gt_500_text1)
print("Number of words ending in 'ed':", num_words_ending_ed_text1)

```
