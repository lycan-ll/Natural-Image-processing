```python
# Corpus names
corpus_names = ["Text1", "Text2", "Text3", "Text4", "Text5", "Text6", "Text7", "Text8", "Text9"]

# Assigning text variables
texts = [text1, text2, text3, text4, text5, text6, text7, text8, text9]

# Corpus information
corpus_info = []

for i, text in enumerate(texts):
    corpus_info.append({
        "Corpus Name": corpus_names[i],
        "Corpus Length": len(text),
        "Unique Words": len(set(text)),
        "Lexical Richness": len(set(text)) / len(text)
    })

# Displaying corpus information
for info in corpus_info:
    print("Corpus:", info["Corpus Name"])
    print("Corpus Length:", info["Corpus Length"])
    print("Unique Words:", info["Unique Words"])
    print("Lexical Richness:", info["Lexical Richness"])
    print()
```

```python
# Tokens to analyze
tokens_to_analyze = ['monster', 'evil', 'devil', 'the']

# Calculate TF, LOGTF, IDF for each token in Text1
text1_values = {}
for token in tokens_to_analyze:
    text1_values[token] = {
        'TF()': text1.count(token) / len(text1) * 100,
        'LOGTF()': math.log(text1.count(token) + 1, 10),
        'IDF()': math.log(len(text1) / (text1.count(token) + 1), 10)
    }

# Get 10 most common words in Text1
fdist1 = FreqDist(text1)
common_words = fdist1.most_common(10)

# Display the table
print("Text1\n")
print("Tokens\t\tTF()\t\tLOGTF()\t\tIDF()")

for token in tokens_to_analyze:
    print(f"{token}\t\t{text1_values[token]['TF()']:.2f}%\t\t{text1_values[token]['LOGTF()']:.2f}\t\t{text1_values[token]['IDF()']:.2f}")

print("\nCommon words in Text1")
for idx, word in enumerate(common_words[:3]):
    print(f"Common word {idx + 1}: {word[0]}")
```

# Task 3: Tokenization & POS

```python
nltk.download('punkt')
text = "NLTK is a powerful library for natural language processing."
words = nltk.word_tokenize(text)
sentences = nltk.sent_tokenize(text)
print(words)
print(sentences)
```

```python
nltk.download('averaged_perceptron_tagger')
tags = nltk.pos_tag(words)
print(tags)
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
filtered_words = [word for word in words if word.lower() not in
stopwords.words('english')]
print(filtered_words)
```
